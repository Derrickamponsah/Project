DATASET VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dataset = pd.read_csv('red_wine.csv')
dataset
```

```python
x = dataset.alcohol
y = dataset.pH
plt.figure(figsize = (7,3))
plt.xlabel('alcohol')
plt.ylabel('pH')
plt.title('A Plot of Alcohol vrs pH')
plt.plot(x,y, marker = 'x')
plt.show()
```

```python
x = dataset.alcohol
y = dataset.pH
plt.figure(figsize = (7,3))
plt.xlabel('alcohol')
plt.ylabel('pH')
plt.title('A Plot of Alcohol vrs pH')
plt.bar(x,y)
plt.show()
```

```python
y1 = dataset.sulphates
x1= dataset.alcohol
plt.ylabel('Sulphate content')
plt.xlabel('Alcohol')
plt.title('A Graph of Sulphate content vrs Alcohol')
plt.bar(x1,y1)
```

```python
plt.scatter(x1,y1, marker = '*')
plt.title('A Scatter diagram of Sulphate content VRS Alcohol')
plt.xlabel('Alcohol')
plt.ylabel('Sulphate content')
```

```python
y2 = dataset.sulphates
x2= dataset.pH
plt.ylabel('Sulphate content')
plt.xlabel('pH')
plt.title('A Graph of Sulphate content vrs pH')
plt.bar(x2,y2)
```

```python
y2 = dataset.sulphates
x2= dataset.pH
plt.ylabel('Sulphate content')
plt.xlabel('pH')
plt.title('A Graph of Sulphate content vrs pH')
plt.plot(x2,y2, marker = 'o')
```
